## 1. The Context
This project analyzes passengers' satisfaction on the Shinkansen Bullet Train in Japan. By examining various travel and survey parameters, we aim to understand what factors contribute to a positive travel experience. The insights could potentially inform improvements in train service quality, focusing on key aspects influencing passenger satisfaction.

## 2. The Objective
The objective is to build a machine learning model that predicts whether a passenger was satisfied (Overall_Experience = 1) or not (Overall_Experience = 0) with their travel experience. Additionally, we aim to identify which features are most impactful in determining passenger satisfaction.

## 3. The Key Questions
- What are the primary factors influencing passenger satisfaction on the Shinkansen Bullet Train?
- How can travel and survey data be effectively used to predict passenger satisfaction?
- Which model performs best in predicting passenger satisfaction, achieving an accuracy of 0.96 or above?

## 4. The Problem Formulation
This project is formulated as a binary classification problem. We use labeled data (Overall_Experience as the target variable) to train a model on various features from both travel and survey data, aiming to predict satisfaction levels for unseen test data. The success criterion is achieving high accuracy in predicting the binary satisfaction outcome (1 for satisfied, 0 for not satisfied).

```python
# Function to calculate and print skewness
def check_skewness(df, numerical_columns):
    for column in numerical_columns:
        skewness = skew(df[column].dropna())  # Drop NaN values to calculate skewness
        print(f"Skewness of {column}: {skewness:.2f}")
        if abs(skewness) > 1:
            print(f"   - {column} is highly skewed\n")
        elif abs(skewness) > 0.5:
            print(f"   - {column} is moderately skewed\n")
        else:
            print(f"   - {column} is approximately symmetric\n")

# Numerical columns in Travel Data
print("Travel Data - Skewness Analysis\n")
numerical_columns_travel = ['Age', 'Travel_Distance', 'Departure_Delay_in_Mins', 'Arrival_Delay_in_Mins']
check_skewness(travel_data, numerical_columns_travel)

# Numerical columns in Survey Data (if any)
print("Survey Data - Skewness Analysis\n")
numerical_columns_survey = ['Overall_Experience']  # Although this is binary, itâ€™s included for completeness
check_skewness(survey_data, numerical_columns_survey)
```

```python
# Step 1: Calculate Gender Distribution by Age Group
age_group_gender_dist = df_copy.dropna(subset=['Gender']).groupby('Age_Group')['Gender'].value_counts(normalize=True).unstack()

# Step 2: Function to Impute Gender Based on Age Group Distribution
def impute_gender(row):
    if pd.isna(row['Gender']):
        if not pd.isna(row['Age_Group']):  # Age group is available
            # Use Age_Group-based probabilities to determine Gender
            gender_dist = age_group_gender_dist.loc[row['Age_Group']]
            return np.random.choice(gender_dist.index, p=gender_dist.values)
        else:
            # If Age Group is also missing, fallback to mode imputation
            return df_copy['Gender'].mode()[0]
    else:
        return row['Gender']

# Step 3: Apply Imputation
df_copy['Gender'] = df_copy.apply(impute_gender, axis=1)
```

Explanation:
age_group_gender_dist calculates the normalized (percentage-based) distribution of Gender within each Age_Group.
The impute_gender function:
Checks if Gender is missing and uses the distribution of Gender within the Age_Group to randomly assign a value based on those probabilities.
If Age_Group is also missing, it imputes Gender with the most frequent value (mode) from the Gender column.

```python
# Step 1: Calculate mean or mode Age for clusters based on specific satisfaction features
age_clusters = df_copy.groupby(['Ease_of_Online_Booking', 'Online_Support', 'Onboard_Entertainment', 'Onboard_Wifi_Service'])['Age'].mean()

# Step 2: Function to Impute Age Based on Satisfaction Features
def impute_age(row):
    if pd.isna(row['Age']):
        # Check if the satisfaction ratings match any existing clusters
        key = (row['Ease_of_Online_Booking'], row['Online_Support'], row['Onboard_Entertainment'], row['Onboard_Wifi_Service'])
        if key in age_clusters.index:
            return age_clusters[key]  # Impute with mean age for the cluster
        else:
            # If no specific cluster, fallback to general mean age
            return df_copy['Age'].mean()
    else:
        return row['Age']

# Step 3: Apply Imputation
df_copy['Age'] = df_copy.apply(impute_age, axis=1)
```

Explanation:
age_clusters creates average age values for each unique combination of the chosen satisfaction features. This forms a baseline for age clusters.
impute_age function:
Tries to match the missing age row to an existing cluster based on satisfaction ratings.
If a match is found, it assigns the mean age of that cluster.
If no match is found, it falls back to the overall mean age.

Option 2: Conditional Imputation Based on Clusters (Using Clustering Algorithm)
This approach clusters the data based on similar values of Seat_Class, Gender, Age_Group, Cleanliness, and Travel_Class, then fills missing Seat_Comfort values with the most frequent value in each cluster. This method is useful when there are distinct patterns in the data based on the selected features.

```python
# Check if missing values align
missing_same_rows = df_copy['Platform_Location'].isna() & df_copy['Onboard_Wifi_Service'].isna()
print("Do Platform_Location and Onboard_Wifi_Service have missing values in the same rows?")
print(missing_same_rows.all())  # This should return True if they align

# Step 1: Impute Platform_Location based on Seat_Class and Travel_Class
df_copy['Platform_Location'] = df_copy.groupby(['Seat_Class', 'Travel_Class'])['Platform_Location']\
                                      .apply(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else None))\
                                      .reset_index(level=[0, 1], drop=True)

# Step 2: Impute Onboard_Wifi_Service based on Seat_Class, Travel_Class, and Overall_Experience
df_copy['Onboard_Wifi_Service'] = df_copy.groupby(['Seat_Class', 'Travel_Class', 'Overall_Experience'])['Onboard_Wifi_Service']\
                                          .apply(lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else None))\
                                          .reset_index(level=[0, 1, 2], drop=True)

# Check remaining missing values after imputation
print("Remaining missing values after group-based mode imputation:")
print(df_copy[['Platform_Location', 'Onboard_Wifi_Service']].isna().sum())
```

In the provided solution, group-based mode imputation was used. Hereâ€™s a breakdown of the approach:

Group-Based Mode Imputation: We grouped the data based on specific columns that were relevant to the features with missing values (Platform_Location and Onboard_Wifi_Service), using insights from previous univariate and bivariate analyses.

For Platform_Location, the data was grouped by Seat_Class and Travel_Class.
For Onboard_Wifi_Service, we used an additional grouping with Overall_Experience (Seat_Class, Travel_Class, and Overall_Experience).
Mode Calculation within Groups: Within each group, the mode (most common value) of Platform_Location and Onboard_Wifi_Service was calculated:

Mode Imputation: If the groupâ€™s mode was available, it was used to fill the missing values within that group. This is particularly effective for categorical data where we assume that the most frequent value in similar cases (e.g., same Seat_Class and Travel_Class) is a likely imputation.
Fallback Handling: If the mode for a particular group was empty (in cases where there were no non-missing values in the group), the code was set up to avoid throwing an error by not filling those specific values.
Why This Method?
Group-based mode imputation is particularly useful for categorical data with missing values because it leverages patterns within subgroups. By focusing on characteristics like Seat_Class and Travel_Class, we made assumptions that similar travel and seating conditions may have similar platform and onboard service attributes, thus providing a contextually accurate fill for missing data.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Define the number of clusters
n_clusters = 5

# Select relevant features for clustering
cluster_features = ['Seat_Class', 'Travel_Class', 'Overall_Experience', 'Platform_Location', 'Onboard_Wifi_Service']

# Create a copy of the data and drop rows with NaNs in cluster_features for training
df_cluster_train = df_copy[cluster_features].dropna()

# Convert categorical columns to numeric using LabelEncoder
label_encoders = {}
for col in cluster_features:
    if df_cluster_train[col].dtype == 'object':
        le = LabelEncoder()
        df_cluster_train[col] = le.fit_transform(df_cluster_train[col])
        label_encoders[col] = le  # Store encoder for future reference if needed

# Apply KMeans on the encoded data
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df_copy['Cluster'] = kmeans.fit_predict(df_cluster_train)

# 1. Fill missing values for Onboard_Entertainment based on cluster mode using pandas mode()
for cluster in range(n_clusters):
    cluster_indices = df_copy[df_copy['Cluster'] == cluster].index
    # Onboard_Entertainment imputation
    mode_entertainment = df_copy.loc[cluster_indices, 'Onboard_Entertainment'].dropna().mode()
    if not mode_entertainment.empty:
        df_copy.loc[cluster_indices, 'Onboard_Entertainment'] = df_copy.loc[cluster_indices, 'Onboard_Entertainment'].fillna(mode_entertainment.iloc[0])

# 2. Fill missing values for Legroom based on cluster mode using pandas mode()
for cluster in range(n_clusters):
    cluster_indices = df_copy[df_copy['Cluster'] == cluster].index
    # Legroom imputation
    mode_legroom = df_copy.loc[cluster_indices, 'Legroom'].dropna().mode()
    if not mode_legroom.empty:
        df_copy.loc[cluster_indices, 'Legroom'] = df_copy.loc[cluster_indices, 'Legroom'].fillna(mode_legroom.iloc[0])

# Drop the temporary 'Cluster' column
df_copy.drop(columns=['Cluster'], inplace=True)
```

Explanation
Separate Cluster-Based Imputation: The code processes Onboard_Entertainment and Legroom separately to account for their differing numbers of missing values.
Cluster Assignment: KMeans is used to assign each row to a cluster based on non-missing values in cluster_features.

### Clustering and Feature Encoding for Imputation

**Goal**: To group similar records into clusters based on shared characteristics. This allows us to identify patterns and fill missing values based on cluster modes (the most frequent value within each cluster), ensuring that missing data is filled in a way that aligns with similar data points.

**Why Clustering?** By clustering similar rows together, we gain a contextual understanding of each row's typical values. This method is especially useful for categorical data or fields where you want to use logical groupings.

- **OneHotEncoding**: Converts categorical variables into a binary format (one-hot encoding) to make them suitable for clustering and modeling. Here, `OneHotEncoder(sparse_output=False)` is used for encoding categorical features without producing sparse matrices.
- **KMeans**: We apply KMeans clustering based on relevant features, which are chosen based on previous Exploratory Data Analysis (EDA). This partitions data into clusters, each representing a set of similar characteristics for later imputation steps.

### Predictive Modeling for Imputation

**Goal**: For features with a large amount of missing data, clustering alone may not capture sufficient accuracy. In such cases, a classification model (Random Forest Classifier) is used to predict missing values based on observed patterns.

**Why Random Forest?** This classifier is suitable for high-dimensional data, handles categorical data well (after encoding), and provides reasonable accuracy without much tuning. It also captures complex interactions between variables, which can enhance accuracy.

- **Training and Testing**: Each classification model is trained on rows with non-missing values. The model is validated using a test split (e.g., 30% for testing), and the accuracy score is reported to ensure the model performs well before imputing missing values.
- **Imputation**: Missing values are predicted by applying the trained model to the data subset with missing values. This helps ensure that predictions align with real data patterns.

### Feature-Wise Imputation Strategy

Each feature (e.g., Customer_Type, Type_Travel, Arrival_Time_Convenient, etc.) is handled separately with either clustering or predictive modeling, depending on the nature of the feature and the number of missing values. Hereâ€™s how each feature is processed:

- **Customer_Type**: Clustering based on travel-related features (e.g., Travel_Distance, Travel_Class). Missing values are then predicted using a Random Forest model.
- **Type_Travel**: Clustered with features like Customer_Type and Overall_Experience, then imputed with a Random Forest model.
- **Arrival_Time_Convenient**: Imputed using mode within clusters based on features like Platform_Location and Ease_of_Online_Booking.
- **Catering**: Filled by mode within clusters created using related features like Onboard_Service and Arrival_Time_Convenient.
- **Onboard_Service**: Imputed with the mode of clusters based on features like Onboard_Entertainment and Onboard_Wifi_Service.

### Final Cleanup

- After imputation, the temporary cluster columns are dropped, as theyâ€™re no longer needed.
- **Final Check for Missing Values**: A summary of remaining missing values is printed, ensuring all columns are fully filled.

### Advantages and Accuracy

- **High Accuracy**: By combining clustering (logical grouping) and predictive modeling, the method minimizes the risk of imputation errors and aligns filled values with similar records.
- **Contextual Imputation**: Each missing value is filled with respect to similar data points, preserving relationships between variables and improving the likelihood of realistic values.
- **Efficient Handling of Complex Data**: This method works well for datasets with mixed categorical and numerical features, ensuring even large proportions of missing data are handled systematically.
